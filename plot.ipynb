```python
def Aktivitaet(t,T12,N0):
    #t= np.arange(0,10*T12,100)
    _lambda = np.log(2)/ T12
    N = N0 * np.exp(-_lambda *t)
    return N

fig = plt.figure()
ax3 = fig.add_subplot(111)



# halbwertzeiten:
T12_C11 = 20.364 * 60
T12_N13 = 9.965 * 60
T12_O14 = 70
T12_O15 = 122.24
T12_F18 = 109.77 *60


t= np.arange(0,10000,10)


#C11
plt.plot(t,Aktivitaet(t,T12_C11,A116[0]))
plt.plot(t,Aktivitaet(t,T12_N13,A137[0]))
plt.plot(t,Aktivitaet(t,T12_O14,A148[0]))
plt.plot(t,Aktivitaet(t,T12_O15,A158[0]))
plt.plot(t,Aktivitaet(t,T12_F18,A189[0]))

plt.plot(t,Aktivitaet(t,T12_C11,A116[0])+
         Aktivitaet(t,T12_N13,A137[0])+
         Aktivitaet(t,T12_O14,A148[0])+
         Aktivitaet(t,T12_O15,A158[0])+
         Aktivitaet(t,T12_F18,A189[0]),label='Summe')
plt.legend()
```
